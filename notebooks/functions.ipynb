# Functions
This notebook just contains some common functions for discussion in class.

```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
x = np.arange(-10., 10., 0.01)
```

```python
y = x**2
pl.plot(x, y, label=r'$x^2$')
pl.legend()
pl.show()
```

```python
y = x**3
pl.plot(x, y, label=r'$x^3$')
pl.legend()
pl.show()
```

```python
y = x**4
pl.plot(x, y, label=r'$x^4$')
pl.legend()
pl.show()
```

```python
y = x**5
pl.plot(x, y, label=r'$x^5$')
pl.legend()
pl.show()
```

```python
y = x**6
pl.plot(x, y, label=r'$x^6$')
pl.legend()
pl.show()
```

```python
y = x**7
pl.plot(x, y, label=r'$x^7$')
pl.legend()
pl.show()
```

```python
y = np.exp(x)
pl.plot(x, y, label=r'$e^2$')
pl.legend()
pl.show()
```

```python
y = np.sqrt(x[x >= 0])
pl.plot(x[x >= 0], y, label=r'$\sqrt{x}$')
pl.legend()
pl.show()
```

```python
y = x**2 + 100
pl.plot(x, y, label=r'$x^2 + 100$')
pl.legend()
pl.show()
```

```python
y = x**2 - 100
pl.plot(x, y, label=r'$x^2 - 100$')
pl.legend()
pl.show()
```

```python
y = x**2 - 5*x + 4
pl.plot(x, y, label=r'$x^2 - 5x + 4$')
pl.legend()
pl.show()
```

```python
y = np.tanh(x)
pl.plot(x, y, label=r'$\tanh(x)$')
pl.legend()
pl.show()
```

```python
y = 1 / (1 + np.exp(-x))
pl.plot(x, y, label=r'$S(x)$')
pl.legend()
pl.show()
```
